```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from pathlib import Path
from matplotlib.lines import Line2D
colors = ["steelblue", "lightcoral", "lightgreen"]
custom_lines = [Line2D([0], [0], color=colors[0], lw=4),
                Line2D([0], [0], color=colors[1], lw=4),
                Line2D([0], [0], color=colors[2], lw=4)]
```

```python
project_dir = str(Path(os.getcwd()).parent)
datasets_path = project_dir + "/datasets/"
df = pd.read_csv(datasets_path + "down_detector_reported_problem_count.csv")
```

```python
# Extract the reasons
def extract_reasons_magnitude(df):
    reasons = dict()
    providers = df["provider"].unique()
    for vendor in providers:
        df_vendor = df[df['provider'] == vendor]
        events = df_vendor["event_id"].unique()

        reason_percent = dict()
        for event in events:
            df_ev = df_vendor[df_vendor["event_id"] == event]
            reports = df_ev["report_id"].unique()

            for report_id in reports:
                df_rep = df_ev[df_ev["report_id"] == report_id]
                causes = df_rep["cause"].values 
                percents = df_rep["percent"].values

            for c, p in zip(causes, percents):
                if c in reason_percent.keys():
                    reason_percent[c] += p
                else:
                    reason_percent[c] = p

        reasons[vendor] = reason_percent
        del reasons[vendor][np.NaN]
        
        vendor_total = sum(reasons[vendor].values())
        for k, v in reasons[vendor].items():
            reasons[vendor][k] = round(100 * (v / vendor_total), 2)
            
    return reasons
```

```python
extracted_reasons = extract_reasons_magnitude(df)
```

```python
# TODO: Make barh, put horizontal line between providers
colors = ["steelblue"] * 4 + ["lightcoral"] * 3 + ["lightgreen"] * 3
fig, ax = plt.subplots(3, 1, figsize=(8, 5), constrained_layout=True, sharex=True)
aws = extracted_reasons['aws']
gcp = extracted_reasons['gcp']
azure = extracted_reasons['azure']

ax[0].barh([k.replace(" ", '\n') for k in list(aws.keys())], list(aws.values()), 0.3, color="steelblue", edgecolor="black")
ax[0].get_xaxis().set_visible(False)
ax[1].barh([k.replace(" ", '\n') for k in list(gcp.keys())], list(gcp.values()), 0.3, color="lightcoral", edgecolor="black")
ax[1].set_ylabel("Most Report Problems", fontsize=14)
ax[1].get_xaxis().set_visible(False)
ax[2].barh([k.replace(" ", '\n') for k in list(azure.keys())], list(azure.values()), 0.3, color="lightgreen", label="Microsoft Azure", edgecolor="black")
ax[2].set_xlabel("%", fontsize=14)
ax[2].set_xlim(0, 100)
ax[0].tick_params(axis='both', labelsize=13)
ax[1].tick_params(axis='both', labelsize=13)
ax[2].tick_params(axis='both', labelsize=13)

ax[0].legend(custom_lines, ['Microsoft Azure', 'Google Cloud', 'Amazon Web Services'], ncol=3, loc='center', bbox_to_anchor=(0.5, 1.14), fontsize=12.3)

plt.savefig(project_dir + "/plots/most_reported_problems_per_event_horizontal.pdf", dpi=100)
```
